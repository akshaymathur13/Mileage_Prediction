# **Describe Data**

**Describe Data for Mileage Prediction Project**

The dataset for the mileage prediction project consists of 398 entries and 9 columns, which represent various attributes of cars. Below is an overview of the data:

1. **mpg (Miles Per Gallon)**: This is the target variable that represents the fuel efficiency of a car. It's a continuous numerical value (float) ranging from 9 to 46.6.
  
2. **cylinders**: This indicates the number of cylinders in the car's engine. It's an integer value, typically between 3 and 8.

3. **displacement**: This refers to the total volume of all the cylinders in the car's engine, measured in cubic inches (float).

4. **horsepower**: This represents the car's power output. While most entries are present, there are a few missing values. It's a continuous float value.

5. **weight**: This is the car's weight in pounds, stored as an integer.

6. **acceleration**: This indicates the time (in seconds) the car takes to accelerate from 0 to 60 miles per hour. It's a continuous float value.

7. **model_year**: The year in which the car model was manufactured, represented as an integer.

8. **origin**: This indicates the region where the car was manufactured. It is a categorical variable, with values such as "usa."

9. **name**: This column holds the car's model name, which is a string (categorical) variable.

The dataset contains a combination of numerical and categorical variables, and one column (horsepower) has a few missing values. The target for prediction is the **mpg** (Miles Per Gallon) value.

# **Define Target Variable y and Feature X**

```python
df.columns
```

```python
y = df['mpg']
```

```python
y.shape
```

```python
X = df[['displacement', 'horsepower', 'weight', 'acceleration']]
```

```python
X.shape
```

```python
X
```

# **Scaling Data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X = ss.fit_transform(X)
```

```python
X
```

```python
pd.DataFrame(X).describe()
```

**After Standardization Mean is Zero and Standard Deviation is One**

# **Train test Split Data**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,  y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state = 2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# **Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

**Mileage  = 23.4 - 1.05DisPlacement - 1.68Horsepower - 4.10Weight - 0.11Acceleration + error**

# **Predict Test Data**

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

# **Model Accuracy**

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

4. **Model Selection**: Different regression models, such as Linear Regression, Decision Trees, or Random Forest, are trained to predict **mpg**. The models are evaluated based on their accuracy using metrics like Mean Squared Error (MSE).

5. **Model Training and Testing**: The dataset is split into training and testing sets. Models are trained on the training data and evaluated on the testing data to ensure their performance generalizes well to unseen data.

6. **Hyperparameter Tuning**: Techniques like cross-validation are used to fine-tune model parameters, improving prediction accuracy.

7. **Conclusion**: The model with the best performance is selected, and its prediction accuracy is discussed. The final model predicts the carâ€™s **mpg** based on input features such as horsepower, weight, and cylinders.
